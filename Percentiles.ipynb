# Percentiles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(100, 15, 10000000)

plt.hist(vals, 1000)
plt.show()
```

```python
np.percentile(vals, 50)
```

```python
np.percentile(vals, 84)
```

```python
np.percentile(vals, 98)
```

```python
np.percentile(vals, 99.9)
```

```python
np.percentile(vals, 90)
```

```python
np.percentile(vals, 20)
```
